## Datos de Inflación

Los datos de inflación utilizados en este análisis fueron obtenidos de la página oficial del Instituto Nacional de Estadística y Censos (INDEC) de Argentina. La fuente de los datos es el Índice de Precios al Consumidor (IPC) con cobertura nacional.

Puedes consultar los datos originales en el siguiente enlace proporcionado por el INDEC:

[Índice de Precios al Consumidor - INDEC](https://www.indec.gob.ar/indec/web/Nivel4-Tema-3-5-31)

Los datos descargados incluyen información sobre la inflación mensual en Argentina, que se ha procesado y analizado para obtener estadísticas trimestrales de inflación.


```python
indec = pd.read_excel(r'C:\Users\mauri\OneDrive\Escritorio\Analisis-de-Mercado-Telefonicas-Arg\data\crudo\sh_ipc_09_24.xls', sheet_name=None)
paginador_i = list(indec.keys())
paginador_i
```

```python
ipc = indec[paginador_i[0]] # Leer la hoja
ipc = ipc.iloc[[4, 8]] # Seleccionar solo la inflacion y mes. 
ipc.columns = ipc.iloc[0] # La primera fila como columnas.
ipc = ipc.T # Transponer el dataframe
ipc.columns = ipc.iloc[0] # La primera fila como columnas.
ipc = ipc[1:] # Quitar la pprimera fila
ipc.index.name = None
ipc.reset_index(drop=True, inplace=True) # Restablecer el índice para que sea numérico
ipc['Total nacional'] = ipc['Total nacional'].apply(lambda x: pd.to_datetime(x).strftime('%Y-%m') if pd.to_datetime(x, errors='coerce') is not pd.NaT else x)
ipc['Total nacional'] = pd.to_datetime(ipc['Total nacional'], errors='coerce')


# Tomar solo hasta Marzo. que es hasta donde llegan los ingresos.
ipc = ipc[ipc['Total nacional'] <= '2024-03-01']
# Calcular la inflacion trimestral
ipc.set_index('Total nacional', inplace=True)
ipc = ipc.resample('QE').sum()
ipc = ipc.reset_index()
ipc['Año'] = ipc['Total nacional'].dt.year
ipc['Trimestre'] = ipc['Total nacional'].dt.quarter
ipc['Nivel general'] = ipc['Nivel general'].astype(float)
ipc = ipc.drop(columns=['Total nacional'])

ipc.head()
```

```python
ingresos.loc[ingresos['Año'] == 2033, 'Año'] = 2023 # Modificar un error del df.
ingresos = pd.merge(ingresos, ipc, on=['Año', 'Trimestre'], how='left') # merge para combinar los DataFrames por Año y Trimestre
ingresos.dropna(subset=['Nivel general'], inplace=True)

ingresos.to_csv('../data/crudo/ingresos.csv',index=False)
```
